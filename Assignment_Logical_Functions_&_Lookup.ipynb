## 1. Write a formula to check if the marks in cell B2 are greater than or equal to 40. If yes, return "Pass", otherwise "Fail".

|--------------|-------|

## 2. A company gives a bonus if sales are above 40,000 and attendance is greater than or equal to 90%. Write a formula to calculate whether the employee will get a bonus or not.

## 4. Find the Department of Employee ID = E103. Also explain which LOOKUP function is used and why?

**Dataset:**

| EmpID | Name  | Department |
|-------|-------|------------|
| E101  | Raju  | IT         |
| E102  | Riya  | HR         |
| E103  | Roshni| Finance    |
| E104  | Sejal | Sales      |

**Answer:**

**Formula:**  
**`=VLOOKUP("E103", A2:C5, 3, FALSE)`**

**Result:** Finance

**Explanation:**  
**VLOOKUP** is used because:  
- We are searching vertically (column-wise).  
- The lookup value (EmpID) is in the **leftmost column** of the table.  
- We need an **exact match** (FALSE).

## 7. Find the salary of EmpID = E105 using INDEX-MATCH Function.

**Dataset:**

| EmpID | Salary |
|-------|--------|
| E104  | 55000  |
| E105  | 72000  |
| E106  | 62000  |
| E102  | 41000  |

**Answer:**

**Formula:**  
**`=INDEX(B2:B5, MATCH("E105", A2:A5, 0))`**

**Result:** 72000

**Why INDEX-MATCH?**  
It is more flexible than VLOOKUP. It works even if the lookup column is not on the left and the table is not sorted.

## 8. Use the XLOOKUP function to find the Department of the employee with Emp ID = E303.

**Dataset:**

| EmpID | Department |
|-------|------------|
| E101  | HR         |
| E303  | IT         |
| E106  | Sales      |
| E301  | Finance    |

**Answer:**

**Formula:**  
**`=XLOOKUP("E303", A2:A5, B2:B5)`**

**Result:** IT

**Advantage of XLOOKUP:**  
- Simpler syntax  
- Can look left or right  
- Supports exact, approximate, and wildcard matches  
- Returns error if not found (can add if_not_found argument)
